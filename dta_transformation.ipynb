```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/employee_productivity_dataset.csv')
```

**Q1.** Handle Missing Values (Basic Cleaning)

```python
# Handling missing values based on specified metrics
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Salary'] = df['Salary'].fillna(df['Salary'].mean())
df['Hours_Worked_Per_Week'] = df['Hours_Worked_Per_Week'].fillna(df['Hours_Worked_Per_Week'].median())
df['Performance_Score'] = df['Performance_Score'].fillna(df['Performance_Score'].mean())

print(df)
```

**Q2. **Label Encoding.


```python
from sklearn.preprocessing import LabelEncoder

le_gender = LabelEncoder()
le_dept = LabelEncoder()

# Fit and transform the target columns
df['Gender'] = le_gender.fit_transform(df['Gender'])
df['Department'] = le_dept.fit_transform(df['Department'])

print(df[['Gender', 'Department']])
```

**Q3.** One-Hot Encoding

```python
# Apply one-hot encoding to specified columns
df_encoded = pd.get_dummies(df, columns=['Work_Mode', 'Location'], dtype=int)

print(df_encoded)
print(f"Total columns now: {df_encoded.shape[1]}")
```

**Q4.** Normalization (Min-Max Scaling)

```python
from sklearn.preprocessing import MinMaxScaler

scaler_minmax = MinMaxScaler()
scaled_features = scaler_minmax.fit_transform(df[['Salary', 'Hours_Worked_Per_Week']])

# View outputs mapped back to a dataframe
df_minmax = pd.DataFrame(scaled_features, columns=['Salary_MinMax', 'Hours_MinMax'])
print(df_minmax)
```

**Q5.** Standardization (Scaling)

```python
from sklearn.preprocessing import StandardScaler

scaler_std = StandardScaler()
# Note: Assuming 'Projects_Completed' matches the column name
std_features = scaler_std.fit_transform(df[['Age', 'Projects_Completed']])

df_std = pd.DataFrame(std_features, columns=['Age_Standardized', 'Projects_Standardized'])
print(df_std)
```

**Q6.** Compare Scaling Methods

```python
# Scale the same feature using both approaches
salary_minmax = MinMaxScaler().fit_transform(df[['Salary']])
salary_std = StandardScaler().fit_transform(df[['Salary']])

# Side-by-side verification
comparison_df = pd.DataFrame({
    'Original_Salary': df['Salary'],
    'MinMax_Scaled': salary_minmax.flatten(),
    'Standard_Scaled': salary_std.flatten()
})
print(comparison_df)
```

**Q7.** Build Preprocessing Pipeline

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler

# Isolate numeric and categorical targets for automation
num_cols = ['Age', 'Salary', 'Hours_Worked_Per_Week', 'Projects_Completed']
cat_cols = ['Work_Mode', 'Location']

# Instantiate the pipeline transforms
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), num_cols),
        ('cat', OneHotEncoder(handle_unknown='ignore'), cat_cols)
    ])

print("Pipeline constructed successfully.")
```

**Q8.** Apply Pipeline

Apply the pipeline on the dataset.

Display:

Transformed dataset

Shape of final dataset

```python
# Execute transformation pipeline on raw or clean data subsets
transformed_data = preprocessor.fit_transform(df)

print("Transformed Matrix:\n", transformed_data)
print("Shape of final dataset:", transformed_data.shape)
```

**Q9.** Conceptual Question

Why is scaling important in Python?

 **ANS=**Distance-based Machine Learning models compute mathematical boundaries using geometric coordinate spaces. Without scaling, variables with high absolute magnitudes (like a $100,000 Salary) will completely dominate metrics against minor scales (like 2-5 projects completed), leading to inaccurate gradient descents or skewed distance profiles.

**Q10.** Conceptual Question

Why do we convert categorical data into numerical form?

**ANS=**Machine Learning models rely on structural linear algebra matrices and mathematical loss optimizations. Algorithms cannot calculate mathematical gradients or distances using plaintext strings like "HR" or "Remote"; changing strings into numerical identifiers makes them readable for numerical computer processors.
